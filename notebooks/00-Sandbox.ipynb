## 1: Data Acquisition & Initialization

The raw dataset is retrieved and stored in the data/raw/ directory.<br>
Added pandas read_csv implementation with basic profiling method

```python
import os
import requests
import pandas as pd

# Define paths according to the standard architecture
DATA_URL = "https://full-stack-assets.s3.eu-west-3.amazonaws.com/M03-EDA/Speed+Dating+Data.csv"
RAW_DATA_DIR = "../data/raw"
RAW_DATA_PATH = os.path.join(RAW_DATA_DIR, "speed_dating_data.csv")

# Create the directory if it does not exist
os.makedirs(RAW_DATA_DIR, exist_ok=True)

# Download the dataset if not already present
if not os.path.exists(RAW_DATA_PATH):
    print("Downloading dataset...")
    response = requests.get(DATA_URL)
    with open(RAW_DATA_PATH, 'wb') as file:
        file.write(response.content)
    print("Download complete.")
else:
    print("Dataset already exists locally.")

# Load the data into a DataFrame
# 'latin1' encoding is required to handle special characters present in the text fields
df = pd.read_csv(RAW_DATA_PATH, encoding="latin1")

# Display initial profiling information
print(f"Dataset Shape: {df.shape}")
display(df.head())
```

## 2: Missing Value Assessment & Baseline Cleaning

Visualizing Missingness: Plotting the percentage of missing values provides an immediate, intuitive understanding of data quality. It quickly highlights which features are viable for analysis and which should be discarded.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")

# Calculate percentage of missing values per column
missing_percent = df.isnull().mean() * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

# Visualize the top 50 columns with the highest missing data percentage
plt.figure(figsize=(15, 6))
sns.barplot(x=missing_percent.index[:50], y=missing_percent.values[:50], palette="viridis")
plt.xticks(rotation=90)
plt.title("Percentage of Missing Values per Feature (Top 50)")
plt.ylabel("% Missing")
plt.xlabel("Features")
plt.tight_layout()
plt.show()

# Implement MVP cleaning strategy: Drop columns with > 50% missing values
threshold = 50.0
cols_to_drop = missing_percent[missing_percent > threshold].index.tolist()
df_cleaned = df.drop(columns=cols_to_drop)

# Output the cleaning results
print(f"Threshold applied: Drop columns with > {threshold}% missing data.")
print(f"Number of columns dropped: {len(cols_to_drop)}")
print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")
```

As a first iteration, features with over 50% missing data are dropped. Attempting to impute a feature where the majority of data is absent introduces severe bias and artificial noise. Dropping them ensures the baseline analysis relies on actual recorded signals. However, some columns need to be "protected" such as identifiers (iid, pid), target variables (match, dec, dec_o), and fundamental demographics (gender) as they represent structural anchors for further analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")

# Calculate percentage of missing values per column
missing_percent = df.isnull().mean() * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

# Define anchor columns that must not be dropped under any circumstances
# iid/pid: subject and partner IDs
# match/dec/dec_o: target variables representing the decision and final match
# gender: fundamental demographic partition
MUST_KEEP = ['iid', 'pid', 'match', 'dec', 'dec_o', 'gender']

# Implement refined MVP cleaning strategy: 
# Drop columns with > 50% missing values UNLESS they are in the MUST_KEEP list
threshold = 50.0
initial_drop_candidates = missing_percent[missing_percent > threshold].index.tolist()

cols_to_drop = [col for col in initial_drop_candidates if col not in MUST_KEEP]
df_cleaned = df.drop(columns=cols_to_drop)

# Output the cleaning results
print(f"Threshold applied: > {threshold}% missing data.")
print(f"Anchor columns protected: {MUST_KEEP}")
print(f"Number of columns actually dropped: {len(cols_to_drop)}")
print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")
```

filtering step to exclude waves 6, 7, 8, and 9 based on the wave column.

```python
import os
import pandas as pd

# 1. Address Data Coherency: Filter out waves 6-9
# These waves used a 1-10 scale instead of a 100-point allocation scale
df_coherent = df[~df['wave'].isin([6, 7, 8, 9])].copy()

# 2. Re-apply Structural Protection and Missing Value Filter (50% threshold)
MUST_KEEP = ['iid', 'pid', 'match', 'dec', 'dec_o', 'gender']
missing_percent = df_coherent.isnull().mean() * 100
initial_drop_candidates = missing_percent[missing_percent > 50.0].index.tolist()
cols_to_drop = [col for col in initial_drop_candidates if col not in MUST_KEEP]

df_cleaned = df_coherent.drop(columns=cols_to_drop)

# 3. Re-apply Baseline Imputation
numeric_cols = df_cleaned.select_dtypes(include=['number']).columns
categorical_cols = df_cleaned.select_dtypes(exclude=['number']).columns

num_imputation = df_cleaned[numeric_cols].median()
cat_imputation = df_cleaned[categorical_cols].mode().iloc[0]

df_cleaned[numeric_cols] = df_cleaned[numeric_cols].fillna(num_imputation)
df_cleaned[categorical_cols] = df_cleaned[categorical_cols].fillna(cat_imputation)

# 4. Update the Processed Data Checkpoint
PROCESSED_DATA_DIR = "../data/processed"
PROCESSED_DATA_PATH = os.path.join(PROCESSED_DATA_DIR, "speed_dating_cleaned.csv")
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)
df_cleaned.to_csv(PROCESSED_DATA_PATH, index=False)

print(f"Waves 6-9 filtered out.")
print(f"New Cleaned DataFrame shape: {df_cleaned.shape}")
print(f"Updated dataset checkpoint saved at: {PROCESSED_DATA_PATH}")
```

### Scale Incompatibility: Mixing a 100-point allocation scale with a 1-10 rating scale invalidates any comparative statistics (like mean or variance).

Scale Consistency: The "in-the-moment" date ratings (attr, sinc, intel, fun, amb, shar) were consistently measured on a 1-10 scale across all waves. Only the stated preference surveys (e.g., attr1_1, distributing points to indicate importance) suffered from the methodology change in waves 6-9

Statistical Power: Filtering out waves 6-9 from the entire dataset unnecessarily discards roughly 20% of the valuable "in-the-moment" behavioral data. Maintaining two tracks is the optimal data science approach. It maximizes statistical power (N) for behavioral analysis while ensuring mathematical validity for preference analysis

Two-Track Architecture: * Track 1 (Behavioral): Uses all waves. Focuses on demographics, decisions (dec), and how the 1-10 partner ratings drive those decisions.

Track 2 (Preferences): Excludes waves 6-9. Focuses on the trade-offs participants stated in their surveys (the 100-point allocations like attr1_1).

## 3: Baseline Imputation & Checkpointing

Separated numerical and categorical columns to apply distinct imputation strategies.

Applied median imputation for numerical data and mode imputation for categorical data.

Created the data/processed/ directory structure.

Exported the finalized baseline dataset to a new CSV file.

```python
import os

# Isolate numerical and categorical columns
numeric_cols = df_cleaned.select_dtypes(include=['number']).columns
categorical_cols = df_cleaned.select_dtypes(exclude=['number']).columns

# Calculate imputation values
num_imputation = df_cleaned[numeric_cols].median()
cat_imputation = df_cleaned[categorical_cols].mode().iloc[0]

# Apply baseline imputation
df_cleaned[numeric_cols] = df_cleaned[numeric_cols].fillna(num_imputation)
df_cleaned[categorical_cols] = df_cleaned[categorical_cols].fillna(cat_imputation)

# Verify no missing values remain
remaining_missing = df_cleaned.isnull().sum().sum()
print(f"Total missing values remaining: {remaining_missing}")

# Define paths for the processed data checkpoint
PROCESSED_DATA_DIR = "../data/processed"
PROCESSED_DATA_PATH = os.path.join(PROCESSED_DATA_DIR, "speed_dating_cleaned.csv")

# Create directory and export the cleaned dataset
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)
df_cleaned.to_csv(PROCESSED_DATA_PATH, index=False)

print(f"Cleaned dataset checkpoint created at: {PROCESSED_DATA_PATH}")
```

## 4: Target Variable Exploration & Gender Distribution
Analyzing the target variable (match) is the foundational step of the Exploratory Data Analysis. It establishes the baseline probability of success and highlights underlying class imbalances

```python
import plotly.express as px
import pandas as pd

# Map gender for clearer visual labels (0 = Female, 1 = Male based on dataset key)
df_cleaned['gender_label'] = df_cleaned['gender'].map({0: 'Female', 1: 'Male'})
df_cleaned['match_label'] = df_cleaned['match'].map({0: 'No Match', 1: 'Match'})

# 1. Global Match Rate Analysis
match_counts = df_cleaned['match_label'].value_counts().reset_index()
match_counts.columns = ['Match Status', 'Count']

fig_global = px.pie(
    match_counts, 
    names='Match Status', 
    values='Count', 
    title='Global Match Rate in Speed Dating Experiment',
    color='Match Status',
    color_discrete_map={'No Match': '#EF553B', 'Match': '#00CC96'},
    hole=0.4
)
fig_global.update_traces(textposition='inside', textinfo='percent+label')
fig_global.show()

# 2. Match Rate by Gender Analysis
# Calculate the percentage of matches within each gender group
gender_match_stats = df_cleaned.groupby(['gender_label', 'match_label']).size().reset_index(name='Count')
gender_totals = df_cleaned.groupby('gender_label').size().reset_index(name='Total')
gender_match_stats = pd.merge(gender_match_stats, gender_totals, on='gender_label')
gender_match_stats['Percentage'] = (gender_match_stats['Count'] / gender_match_stats['Total']) * 100

fig_gender = px.bar(
    gender_match_stats, 
    x='gender_label', 
    y='Percentage', 
    color='match_label',
    title='Match Distribution by Gender',
    labels={'gender_label': 'Gender', 'Percentage': 'Percentage of Total Dates (%)'},
    color_discrete_map={'No Match': '#EF553B', 'Match': '#00CC96'},
    barmode='stack',
    text=gender_match_stats['Percentage'].apply(lambda x: f'{x:.1f}%')
)
fig_gender.update_layout(xaxis_title="Gender", yaxis_title="Percentage (%)")
fig_gender.show()
```

### Visualisation Interpretations: Target Variable & Gender

* **Global Match Rate:** The global match rate is highly imbalanced, with actual matches occurring in less than 20% of the speed dates. This confirms that finding mutual interest is a rare event, which perfectly aligns with the marketing team's observation of decreasing match volumes.
* **Gender Distribution:** When broken down by gender, the stacked bar chart indicates whether males or females experience different baseline success rates. (Note: The interaction rate might appear similar here because a "match" requires both parties to say yes, meaning the absolute number of matches is identical for both genders. The true behavioral difference lies in the individual *decisions* (`dec`), not the final mutual `match`).

## 5: Individual Decision Rates by Gender

```python
import plotly.express as px
import pandas as pd

# Map the individual decision for clearer visual labels (0 = No, 1 = Yes)
df_cleaned['dec_label'] = df_cleaned['dec'].map({0: 'No', 1: 'Yes'})

# Calculate individual decision rates grouped by gender
dec_stats = df_cleaned.groupby(['gender_label', 'dec_label']).size().reset_index(name='Count')
gender_totals_dec = df_cleaned.groupby('gender_label').size().reset_index(name='Total')

# Merge to calculate percentages
dec_stats = pd.merge(dec_stats, gender_totals_dec, on='gender_label')
dec_stats['Percentage'] = (dec_stats['Count'] / dec_stats['Total']) * 100

# Visualize the decision distribution by gender
fig_dec = px.bar(
    dec_stats, 
    x='gender_label', 
    y='Percentage', 
    color='dec_label',
    title='Individual Decision Rates by Gender ("Swipe Right")',
    labels={
        'gender_label': 'Gender', 
        'Percentage': 'Percentage of Decisions (%)', 
        'dec_label': 'Decision'
    },
    color_discrete_map={'No': '#EF553B', 'Yes': '#00CC96'},
    barmode='stack',
    text=dec_stats['Percentage'].apply(lambda x: f'{x:.1f}%')
)

fig_dec.update_layout(xaxis_title="Gender", yaxis_title="Percentage (%)")
fig_dec.show()
```

### Visualisation Interpretations: Individual Decisions (`dec`)

* **Behavioral Differences:** The visualization reveals a clear asymmetry in dating behavior between genders. Males express interest ("Yes") at a significantly higher rate than females. 
* **Tinder Translation:** In the context of the Tinder app, this indicates that men "swipe right" much more frequently, acting as the primary drivers of potential match volume, while women are generally more selective. This selective behavior acts as the bottleneck for the overall global match rate.
